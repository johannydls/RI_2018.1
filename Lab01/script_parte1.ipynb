## Importações necessárias para a análise dos dados

```python
import pandas as pd
import nltk
```

## Definindo o arquivo com os dados a ser analisado

```python
dataset = pd.read_csv('../data/noticias_estadao.csv', sep=',')
```

## Realizando junção dos títulos e conteúdos

```python
documentos = dataset.titulo + " " + dataset.conteudo
```

## Criação dos docIDs

```python
docIDs = dataset.idNoticia
```

```python
indiceInvertido = {}
```

## Fazendo download de biblioteca necessária para realizar a tokenização das palavras

```python
nltk.download('punkt')
```

## Criando os tokens dos dados

```python
tokens = documentos.apply(nltk.word_tokenize)
```

## Função geradora do índice invertido

```python
def geraIndiceInvertido():
    for i in range(len(tokens)):
        idNoticia = docIDs[i]
        palavras = tokens[i]
    
        for palavra in palavras:
            if palavra.lower() not in indiceInvertido:
                indiceInvertido[palavra.lower()] = [idNoticia]
        
            elif idNoticia not in indiceInvertido[palavra.lower()]:
                indiceInvertido[palavra.lower()].append(idNoticia)
```

```python
geraIndiceInvertido()
```

## Funções de Busca Booleana

```python
def busca_1_termo(termo):
    return indiceInvertido[termo.lower()] 
```

```python
def busca_AND(termo1, termo2):

    intersecao = set(indiceInvertido[termo1.lower()])
    
    ids = set(indiceInvertido[termo1.lower()])
    intersecao = intersecao.intersection(ids)
    
    ids = set(indiceInvertido[termo2.lower()])
    intersecao = intersecao.intersection(ids)
        
    return list(intersecao)
```

```python
def busca_OR(termo1, termo2):
    
    uniao = set(indiceInvertido[termo1.lower()])
    
    ids = set(indiceInvertido[termo1.lower()])
    uniao = uniao.union(ids)
    
    ids = set(indiceInvertido[termo2.lower()])
    uniao = uniao.union(ids)
        
    return list(uniao)
```

## Função genérica para realizar buscas de qualquer tipo

```python
def search(parametro):
    
    entradas = parametro.split(' ')
    
    if (len(entradas) >= 2):
        termo1 = entradas[0]
        termo2 = entradas[2]
        operador = entradas[1]
        
        if (operador.lower() == 'and'):
            return busca_AND(termo1, termo2)
        
        elif (operador.lower() == 'or'):
            return busca_OR(termo1, termo2)

    elif (len(entradas) == 1):
        return busca_1_termo(entradas[0])
    
    else:
        print("Apenas buscas 1-termo, AND e OR")
        return
```

## Função para retornar as palavras mais próximas da pesquisada

```python
def busca_aproximada(termo):
    termos = indiceInvertido.keys()
    return sorted(termos, key=lambda palavra: nltk.edit_distance(termo,palavra))[0:3]
```

```python
busca_aproximada('poiltica')
```

# Realizando as consultas solicitadas

## Busca: Campina AND Grande

```python
def sanity_check():
    busca = sorted(search("Campina AND Grande"))
    esperado = [1952, 4802, 1987, 6694, 5382, 1770, 2763, 1068, 5870, 2777, 1370, 2779]
    
    return sorted(busca) == sorted(esperado)
```

```python
sanity_check()
```

### Validação

```python
assert len(search("Campina AND Grande")) == 12
search("Campina AND Grande")
```
